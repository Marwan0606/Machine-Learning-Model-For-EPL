```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
print("It worked!")
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1) #the higher the n-estimators the longer the model will take but it potentially be more accurate. #n_estimators is the number of individual decision trees we wanna train. Random forest is a series of decision trees that each decision trees has slightly different parameters.
#min_samples_split is the number of samples we want to have in a leaf of the decision tree before we split the node, the higher the value the less likely you are to overfit but the lower the accuracy may be #random_state is like the seed or type of order you need from the model to get the same result, so if I set my random_state to 1 then I will always get the same results but if I choose 2 I will get different results but if I choose 2 again I will get the same results for the random_state 2
#parameters are like the questions the models asks the data to get a yes or no
```

```python
train = matches[matches["date"] < '2022-01-01'] #so we are making our training data and we NEED TO MAKE SURE THE TRAINING DATA IS FROM BEFORE THE TEST DATA. We always want to predict the future from the past not the opposite
```

```python
test = matches[matches["date"] > '2022-01-01']
```

```python
#we split our train and test set because if we dont and just train our model on the same dataset we will test it on then our model will seem to be very accurate, even though the model isnt that accurate with new data. This is like teacher giving you the same exact test as the practice test you did at home the day before
```

```python
predictors = ["venue_code", "opp_code", "hour", "day_code"] #list of predictor columns we created earlier
```

```python
rf.fit(train[predictors], train["target"]) #this line trains a random forest model using the predictors to try to predict the target
```

```python
preds = rf.predict(test[predictors]) #this generates predictions using the predict method where we pass on our test data
```

```python
#Now what we want is to have rolling averages were we group each team so we can see how well they did the last few matches, how many goals, how many goals they conceded...
```

```python
grouped_matches = matches.groupby("team") #creating a dataframe which groups matches by team, so this will create one dataframe for every team
```

```python
group = grouped_matches.get_group("Manchester City") #this gives us all the matches manchester city played in
```

```python
# what we want to do is if we are for ex in match week 10 then how did manchester city do in their last 3 games (matchweek 7,8,9) and then we can pass that into algorithm so algorithm can make better predictions
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date") #sorting by ascending order by date
    rolling_stats = group[cols].rolling(3, closed="left").mean() #take a set of columns cols and compute the average for them, closed = "left" ensures when we for example calculated the averages for matchweek 7,8,9 we then use those averages values for predicting week 10 not week 9, if we use it to predict week 9 instead of 10 then we are using information from the future which is cheating and will make our model look better than it actualy is
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols) #this is saying if their are missing values then remove all those rows with missing values, this happens for example when we are trying to predict matchweek 2 so there is only 1 matchweek before we can use and not 3
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"] #these are the columns we wanna calculate rolling averages for
new_cols = [f"{c}_rolling" for c in cols] #here just defining new columns to put our rolling averages values in
```

```python
new_cols #new_cols are the columns with new averages in them
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols)) #what apply does is apply the rolling_averages function we created to each team 
```

```python
def make_predictions(data, predictors): #WE ALREADY DID EVERYTHING IN THIS FUNCTION RIGHT NOW WE ARE JUST COMBINING ALL THOSE STUFF INTO THIS ONE NICE FUNCTION
    train = data[data["date"] < '2022-01-01'] #so we are making our training data and we NEED TO MAKE SURE THE TRAINING DATA IS FROM BEFORE THE TEST DATA. We always want to predict the future from the past not the opposite
    test = data[data["date"] > '2022-01-01']
    rf.fit(train[predictors], train["target"]) #this line trains a random forest model using the predictors to try to predict the target
    preds = rf.predict(test[predictors]) #this generates predictions using the predict method where we pass on our test data
    combined = pd.DataFrame(dict(actual=test["target"], prediction=preds), index=test.index) #Now we are creating a dataFrane that combines our predicted and actual values
    precision = precision_score(test["target"], preds)
    return combined, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
```

```python
#IMPORTANT SO RIGHT NOW WITH HOW EVERYTHING IS SETUP IT IS VERY POSSIBLE THAT OUR MODEL PREDICTED THAT ARSENAL VS BRENTFORD FOR EXAMPLE ARSENAL WON, AND THEN FOR THAT SAME MATCH BUT BRENTFORD VS ARSENAL THE MODEL COULD HAVE PREDICTED BURNLEY WOULD'VE WON. The problem is that those are both the exact same games only difference is the team and opponent are switched. SO WE NEED TO PROTECT AND STOP THIS FROM HAPPENING
```

```python
#first addressing the problem were in this data we can see the names for team name and opponent name are not exactly the same, so for example here the team name calls the team wolverhampton wanderers but then in the opponents section it calls that same team wolves. #so thats why here we are first normalizing the names
#we use a dictionary to do this and use pandas map function in that dictionary
```

```python
class MissingDict(dict): #here we are making a class that inherits from the dictionary class # what this class does is if someone calls this dictionary and passes on liverpool then instead of throwing an error since liverpool isnt here, instead it will just pass back liverpool
    __missing__ = lambda self, key: key
    
map_values = {
    "Brighton and Hove Albion": "Brighton", #basically saying if the name is Brighton and Hove Albion then just call it Brighton instead
    "Manchester United": "Manchester Utd",
    "Newcastle United": "Newcastle Utd",
    "Tottenham Hotspur": "Tottenham",
    "West Ham United": "West Ham",
    "Wolverhampton Wanderers": "Wolves"
}
mapping = MissingDict(**map_values) #creating an instance of MissingDict
```

```python
mapping["Liverpool"] 
```

```python
mapping["Wolverhampton Wanderers"]
```

```python
combined["new_team"] = combined["team"].map(mapping) #creates a new_team field in the combined table and it does this by taking the name from the team column and running the mapping instance to normalize the name and put it in the new column
```

```python
combined
```

```python
#What the code right under this is saying: look for the new team field and merge that with the opponent field on the other side, I think we are also adding the date to make sure this merging doesnt happen for another game because remember in premier league each team plays against each other twice
```

```python
merged = combined.merge(combined, left_on=["date", "new_team"], right_on=["date", "opponent"]) #merging the dataframe with itself, because what happens is there is the same match but it appears twice (basically each match appears twice) because for example we have once team is arsenal and opponent is burnley but we also have team burnley and opponent is arsenal. these 2 matches are the same but the model does not know that
```
